```python
from pathlib import Path

DATA_PATH = Path("../data/raw/shl_product_catalog.json")

with open(DATA_PATH, "r", encoding="utf-8") as f:
    text = f.read()

print("Characters:", len(text))
print("Lines:", len(text.splitlines()))
```

```python
print(text.count('"entity_id"'))
```

```python
fields = [
    "entity_id",
    "name",
    "description",
    "duration",
    "languages",
    "job_levels",
    "remote",
    "adaptive",
    "link",
    "keys"
]

for field in fields:
    count = text.count(f'"{field}"')
    print(f"{field:<15} -> {count}")
```

```python
broken_names = []

for i, line in enumerate(text.splitlines()):
    if '"name": "' in line and not line.strip().endswith('",'):
        broken_names.append((i + 1, line))

print(f"Broken name lines found: {len(broken_names)}")

for item in broken_names[:20]:
    print(item)
```
